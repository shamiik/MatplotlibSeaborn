* Seaborn also a visualization Library


```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri= sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.tail()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

**Grouping data with Hue**

```python
#sns.lineplot(x='timepoint',y='signal',data=fmri, hue='subject')
#sns.lineplot(x='timepoint',y='signal',data=fmri,hue = 'region')
sns.lineplot(x='timepoint',y='signal',data=fmri,hue = 'event')
plt.show()
```

*`Adding Styles in Data & Add Markers`*


```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue = 'region',style='region',markers=True)
plt.show()
```
