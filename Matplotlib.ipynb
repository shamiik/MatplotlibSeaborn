**Data Visualization**

```python
import numpy as np
from matplotlib import pyplot as plt

x= np.arange(1,11)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.title("First Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color="red",linestyle=":",linewidth=3)
plt.title("First Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
