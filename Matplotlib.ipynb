**Data Visualization**

```python
import numpy as np
from matplotlib import pyplot as plt

x= np.arange(1,11)
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.title("First Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color="red",linestyle=":",linewidth=4)
plt.title("First Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**Adding two lines in the same plot**

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
y1
```

```python
y2
```

```python
plt.plot(x,y1,color='red',linestyle=':',linewidth=2)
plt.plot(x,y2,color='g',linestyle='-.',linewidth=3)
plt.title("2 Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

**SUBPLOT**

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='red',linestyle='-.',linewidth=3)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(2,1,2)
plt.plot(x,y2,color='red',linestyle='-.',linewidth=3)

plt.show()
```

**BAR Plot**

```python
student={'sam': 98,'ptak': 62,'bib':90}
name =list(student.keys())
marks= list(student.values())

#plt.plot(name,marks)
plt.bar(name,marks)


plt.title("Bar Plot")
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

**Horizontal Bar Plot**

```python
plt.barh(name,marks,color="Orange" )
plt.title("Horizontal Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

**Scatter Plot**

```python
p=(1,2,3,4,5,6,7,8,9)
a=(2,6,5,3,9,1,7,4,8)

plt.scatter(p,a)
plt.show()
```

```python
#MARKER and color change
plt.scatter(p,a,marker='*',color='green',s=100)  #s indicates size
plt.show()
```

```python
#Two types of marker
x=(10,20,30,40,50,60,70,80,90)
a=(5,8,9,6,7,3,1,2,4)
b=(2,1,4,3,9,5,7,6,8)
plt.scatter(x,a,color='g',marker='*',s=100)
plt.scatter(x,b,c='r',marker='.',s=80)
plt.show()
```
